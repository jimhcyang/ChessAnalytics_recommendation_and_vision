```python
from chesscog.recognition import ChessRecognizer
import cv2
import chess
import pickle
import pathlib
import matplotlib.pyplot as plt
```

```python

input_photo = cv2.imread("data/G001_IMG150.jpg")
input_photo = cv2.cvtColor(input_photo, cv2.COLOR_BGR2RGB)

path_models = pathlib.Path("models/transfer_learning_models")
recognizer = ChessRecognizer(path_models)
detected_board, *_ = recognizer.predict(input_photo)
```

```python
plt.imshow(input_photo)
plt.show()

detected_board
```

```python
with open('data/output/chessboard_sample.pkl', 'wb') as f:
    pickle.dump(detected_board, f)
```
