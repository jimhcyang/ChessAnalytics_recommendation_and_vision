```python
# add to python path to allow for chesscog import
import sys
import os
chesscog_dir = os.path.abspath(os.path.join(os.getcwd(), '..', 'chesscog'))
if chesscog_dir not in sys.path:
    sys.path.insert(0, chesscog_dir)
```

```python
from chesscog.recognition import ChessRecognizer
import cv2
import pathlib
import matplotlib.pyplot as plt
```

```python

def detect_board_state(path_image):
    input_photo = cv2.imread(path_image)
    input_photo = cv2.cvtColor(input_photo, cv2.COLOR_BGR2RGB)
    path_models = pathlib.Path(os.path.join(os.path.abspath('..'), 'chesscog', 'models', "transfer_learning_models"))
    recognizer = ChessRecognizer(path_models)
    detected_board, *_ = recognizer.predict(input_photo)
    plt.imshow(input_photo)
    plt.show()
    print(detected_board)
    return detected_board

def get_raw_data_file(file_name):
    file_path = os.path.join(os.path.abspath('..'), 'data', 'raw', file_name)
    return file_path
```

```python

board_1 = detect_board_state(get_raw_data_file("test1.jpg"))
print(f"FEN string notation: {board_1.fen()}")
board_1
```
